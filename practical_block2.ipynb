## EDA block 3

Data comes from Kaggle and shows hourly wages in the U.S. over time by gender and education, from 1973 to 2022. 
Source: https://www.kaggle.com/datasets/asaniczka/usa-wage-comparison-for-college-vs-high-school/

```python
# load the data
import pandas as pd
df = pd.read_csv("~/college_wage_premium.csv")
```

```python
df.head()
```

```python
# check for missing data
df.isna().sum()
```

```python
df.isnull().sum()
```

```python
# plot
import matplotlib
%matplotlib inline
# create a quick bar chart
# looks at different education levels and wage over time
df.plot(x= 'year', y=('high_school'), kind='bar');
df.plot(x= 'year', y='bachelors_degree', kind='bar');
```

```python
# create plots that are easier to read
import matplotlib.pyplot as plt

plt.plot(df['year'], df['high_school'], label = 'high_school')
plt.plot(df['year'], df['bachelors_degree'], label = 'bachelors_degree')
plt.legend()
plt.show()
```

```python
# explore gender differences in highschool
plt.plot(df['year'], df['men_high_school'], label = 'M_highschool')
plt.plot(df['year'], df['women_high_school'], label = 'F_highschool')
plt.legend()
plt.show()
```

```python
# explore gender differences in bachelor's degree
plt.plot(df['year'], df['men_bachelors_degree'], label = 'M_bachelors_degree')
plt.plot(df['year'], df['women_bachelors_degree'], label = 'F_bachelors_degree')
plt.legend()
plt.show()
```

```python
# look at gender differences in wages over time across education levels 

plt.plot(df['year'], df['men_bachelors_degree'], label = 'M_bachelors_degree')
plt.plot(df['year'], df['women_bachelors_degree'], label = 'F_bachelors_degree')
plt.plot(df['year'], df['men_high_school'], label = 'M_highschool')
plt.plot(df['year'], df['women_high_school'], label = 'F_highschool')
plt.legend()
plt.show()
```
